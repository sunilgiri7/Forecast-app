```python
import pandas as pd
from fbprophet import Prophet
import numpy as np
import pystan
```

```python
df = pd.read_csv("sample _forecast _data.csv")
df.head()
```

```python
df[['date', 'value']] = df['date;value'].str.split(';', expand=True)
```

```python
df.drop('date;value', axis=1, inplace=True)
```

```python
df['value'] = pd.to_numeric(df['value'])
```

```python
df.tail(12)
```

```python
import matplotlib.pyplot as plt
plt.plot(df['date'], df['value'])
plt.show()
```

```python
from matplotlib import pyplot
df.plot()
pyplot.show()
```

```python
df.columns = ['ds', 'y']
df['ds'] = pd.to_datetime(df['ds'])
```

```python
model = Prophet(daily_seasonality=True)
model.fit(df)
```

```python
from pandas import read_csv
from pandas import to_datetime
future = list()
for i in range(1,13):
    date = '2017-%20d' % i
    future.append([date])
future = pd.DataFrame(future)
future.columns = ['ds']
future['ds'] = to_datetime(future['ds'])
```

```python
forecast = model.predict(future)
```

```python
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
```

```python
model.plot(forecast)
pyplot.show()
```

```python
# Now predict future
future = list()
for i in range(1,13):
    date = '2018-%20d' % i
    future.append([date])
future = pd.DataFrame(future)
future.columns = ['ds']
future['ds'] = to_datetime(future['ds'])

forecast = model.predict(future)

print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())

model.plot(forecast)
```
